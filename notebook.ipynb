# Bigram

```python
dictionary = {}

for n in names : 
    chs = ['<S>'] + list(n) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        bigram = (ch1, ch2)
        dictionary[bigram]=dictionary.get(bigram, 0) + 1

```

```python
sorted(dictionary.items(), key= lambda kv: -kv[1])
```

```python
import torch
counts = torch.zeros((28, 28), dtype=torch.int32)
```

```python
chars = sorted(list(set(''.join(names))))
stoi = {s:i for i,s in enumerate(chars)}
stoi['<S>']=26
stoi['<E>']=27
```

```python
stoi
```

```python

for n in names : 
    chs = ['<S>'] + list(n) + ['<E>']
    for ch1, ch2 in zip(chs, chs[1:]):
        ix1 = stoi[ch1]
        ix2 = stoi[ch2]
        counts[ix1, ix2]+=1
```

```python
itos = {i:s for s,i in stoi.items()}
```

```python
itos
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,16))
plt.imshow(counts, cmap='Blues')
for i in range(28):
    for j in range(28):
        chstr = itos[i] + itos[j]
        plt.text(j, i, chstr, ha="center", va="bottom")
        plt.text(j, i, counts[i, j].item(), ha="center", va="top")

plt.axis('off');
```
